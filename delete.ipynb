```python
import pandas as pd
```

```python
# load csv as dataframe
df = pd.read_csv("data/wide_to_long_inverters.csv")
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%d/%m/%y %H:%M')
```

```python
df.to_csv("data/wide_to_long_inverters_processed.csv")
```

```python
grouped_data = df.groupby(['Inverter_ID', df['timestamp'].dt.to_period('M')]).agg({'ACTIVE POWER (kW)': 'sum'}).reset_index()
grouped_data
```

```python
filtered_data = grouped_data[grouped_data['Inverter_ID'].str.startswith('January 2024')]
filtered_data
```
