```python
import pandas as pd
```

#1. preprocess `wide_to_long_inverters`

```python
df_raw = pd.read_csv("data/wide_to_long_inverters.csv")
```

```python
df_raw['timestamp'] = pd.to_datetime(df_raw['timestamp'], format='%d/%m/%y %H:%M')
```

```python
# Create a new 'date' column (only the date part)
df_raw['date'] = df_raw['timestamp'].dt.date
```

```python
df_raw.columns
```

```python
# Aggregate by date and compute mean of all numeric columns
df_new = df_raw.drop(columns='timestamp').groupby(['date', 'Inverter_ID']).mean(numeric_only=True).reset_index()
```

```python
df_new.to_csv("data/wide_to_long_inverters_processed.csv", index=False)
```

#2. preprocess `Raw data spansafe for CB.csv`

```python
df = pd.read_csv("data/Raw data spansafe for CB.csv")
```

```python
df
```

```python
# convert wide to long format
df_long = pd.melt(
    df,
    id_vars=['Date', 'Time'],         # ðŸ‘ˆ Do not melt these
    var_name='Column',
    value_name='Values'
)
```

```python
# split "Column" by "-" into multiple columns
df_long[['company', 'device_name', 'value_type']] = df_long['Column'].str.split(' - ', expand=True)
```

```python
# only keey rows that "Device_id" column contains 'Comb String Box'
df_long = df_long[df_long['device_name'].str.contains('Comb String Box', na=False)]
```

```python
df_long[['inverter_id', 'comb_box_ind']] = df_long['device_name'].str.extract(r'([A-Z]{2}\d+)-(\d{2})')
```

```python
# drop columns 'Column' and 'company'
df_long = df_long.drop(columns=['company', 'Column'])
```

```python
# get average values by Date and drop Time column
df_avg = df_long.drop(columns='Time').groupby(['Date', 'device_name', 'value_type', 'inverter_id',
       'comb_box_ind']).mean(numeric_only=True).reset_index()
```

```python
# drop column 'value_type' and rename column 'Values' to 'DC CURRENT(A)'
df_final = df_avg.drop(columns='value_type').rename(columns={'Values': 'DC CURRENT(A)'})
```

```python
df_final
```

```python
df_final.to_csv("data/combiner_box_preprocessed.csv", index=False)
```
